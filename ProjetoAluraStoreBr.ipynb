### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
# Faturamento por loja
print(f'Loja 1 faturou: R${loja["Preço"].sum():.2f}')
print(f'Loja 2 faturou: R${loja2["Preço"].sum():.2f}')
print(f'Loja 3 faturou: R${loja3["Preço"].sum():.2f}')
print(f'Loja 4 faturou: R${loja4["Preço"].sum():.2f}')
```

```python
faturamentos = [loja["Preço"].sum(), loja2["Preço"].sum(), loja3["Preço"].sum(), loja4["Preço"].sum()]
```

# 2. Vendas por Categoria


```python
print(loja.columns)
```

```python
# Top categorias mais vendidas usando o agrupamento
# Loja 1
print("Loja 1 - Quantidade de vendas por Categoria\n")
print(loja.groupby("Categoria do Produto").size().sort_values(ascending=False).to_string())

# Loja 2
print("\nLoja 2 - Quantidade de vendas por Categoria\n")
print(loja2.groupby("Categoria do Produto").size().sort_values(ascending=False).to_string())

# Loja 3
print("\nLoja 3 - Quantidade de vendas por Categoria\n")
print(loja3.groupby("Categoria do Produto").size().sort_values(ascending=False).to_string())

# Loja 4
print("\nLoja 4 - Quantidade de vendas por Categoria\n")
print(loja4.groupby("Categoria do Produto").size().sort_values(ascending=False).to_string())
```

# 3. Média de Avaliação das Lojas

```python
# Média de avaliação por cada loja
print(f'Loja 1 Média de Avaliação: {loja["Avaliação da compra"].mean():.2f}')
print(f'Loja 2 Média de Avaliação: {loja2["Avaliação da compra"].mean():.2f}')
print(f'Loja 3 Média de Avaliação: {loja3["Avaliação da compra"].mean():.2f}')
print(f'Loja 4 Média de Avaliação: {loja4["Avaliação da compra"].mean():.2f}')
```

```python
lojas = ['loja 1', 'loja 2', 'loja 3', 'loja 4']
medias = [loja["Avaliação da compra"].mean(), loja2["Avaliação da compra"].mean(), loja3["Avaliação da compra"].mean(), loja4["Avaliação da compra"].mean()]
```

# 4. Produtos Mais e Menos Vendidos

```python
loja.head()
```

```python
# Top 10 produtos mais vendidos em cada loja
# Loja 1
print("Loja 1 - Quantidade de vendas por Categoria\n")
print(loja.groupby("Produto").size().sort_values(ascending=False).head(10).to_string())

# Loja 2
print("\nLoja 2 - Quantidade de vendas por Categoria\n")
print(loja2.groupby("Produto").size().sort_values(ascending=False).head(10).to_string())

# Loja 3
print("\nLoja 3 - Quantidade de vendas por Categoria\n")
print(loja3.groupby("Produto").size().sort_values(ascending=False).head(10).to_string())

# Loja 4
print("\nLoja 4 - Quantidade de vendas por Categoria\n")
print(loja4.groupby("Produto").size().sort_values(ascending=False).head(10).to_string())
```

# 5. Frete Médio por Loja


```python
# Média de avaliação por cada loja
print(f'Loja 1 Média do Frete: R${loja["Frete"].mean():.2f}')
print(f'Loja 2 Média do Frete: R${loja2["Frete"].mean():.2f}')
print(f'Loja 3 Média do Frete: R${loja3["Frete"].mean():.2f}')
print(f'Loja 4 Média do Frete: R${loja4["Frete"].mean():.2f}')
```

```python
fretes = [loja["Frete"].mean(), loja2["Frete"].mean(), loja3["Frete"].mean(), loja4["Frete"].mean()]
```

# 6. Gerando gráficos




```python
import matplotlib.pyplot as plt
# Avaliação média de cada loja
for i, media in enumerate(medias):
    plt.text(i, media-0.5, f'{media:.2f}', ha='center')

plt.title('Avaliação média de cada loja')
plt.ylabel('Média')
cores = ['blue', 'orange', 'green', 'red']
plt.bar(lojas, medias, color=[cores[i % 4] for i in range(len(medias))])
plt.show()
```

```python
# Faturamento percentual de cada loja
import matplotlib.pyplot as plt

# Criar gráfico
plt.figure(figsize=(7, 7))
plt.pie(faturamentos,labels=lojas,autopct='%.2f%%')
plt.title('Faturamento de cada loja')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Supondo que 'lojas' e 'medias' já estejam definidos
cores = ['blue', 'orange', 'green', 'red']

plt.barh(lojas, fretes, color=[cores[i % 4] for i in range(len(medias))])

# Adicionando os valores ao lado das barras
for i, frete in enumerate(fretes):
    plt.text(frete-5, i, f'R${frete:.2f}', va='center')

plt.title('Valor médio do frete de cada loja')
plt.xlabel(' Valor Frete')
plt.ylabel('Lojas')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Converter a coluna 'Data da Compra' para datetime em cada DataFrame
for df in [loja, loja2, loja3, loja4]:
    df['Data da Compra'] = pd.to_datetime(df['Data da Compra'], errors='coerce')
    df.dropna(subset=['Data da Compra'], inplace=True)


# Agrupar por mês e somar o faturamento
loja1_mes = loja.groupby(pd.Grouper(key='Data da Compra', freq='ME'))['Preço'].sum()
loja2_mes = loja2.groupby(pd.Grouper(key='Data da Compra', freq='ME'))['Preço'].sum()
loja3_mes = loja3.groupby(pd.Grouper(key='Data da Compra', freq='ME'))['Preço'].sum()
loja4_mes = loja4.groupby(pd.Grouper(key='Data da Compra', freq='ME'))['Preço'].sum()

# Criar gráfico com dados mensais
plt.figure(figsize=(10, 6))
plt.plot(loja1_mes.index, loja1_mes.values, label='Loja 1', marker='o')
plt.plot(loja2_mes.index, loja2_mes.values, label='Loja 2', marker='o')
plt.plot(loja3_mes.index, loja3_mes.values, label='Loja 3', marker='o')
plt.plot(loja4_mes.index, loja4_mes.values, label='Loja 4', marker='o')

plt.title('Faturamento mensal por loja')
plt.xlabel('Data')
plt.ylabel('Faturamento (R$)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

## Relatório Final: Análise de Desempenho das Lojas

### Introdução

O objetivo deste relatório é analisar o desempenho de diversas lojas administradas pelo Senhor João, com o intuito de identificar qual delas apresenta o menor potencial de lucratividade e, portanto, deve ser vendida. Para isso, foram avaliados diversos indicadores estratégicos, incluindo o **faturamento total**, **categorias de produtos mais e menos vendidos**, **avaliações médias dos clientes**, **produtos com maior e menor saída**, **frete médio por loja** e o **faturamento mensal por loja**. Algumas análises foram acompanhadas de visualizações gráficas para facilitar a compreensão dos dados.

---

### Desenvolvimento

#### 1. Faturamento Total das Lojas

O gráfico de faturamento mostrou que a **Loja 1** é responsável pela maior parte do faturamento, seguida pela **Loja 2**, enquanto a **Loja 4** apresenta o menor desempenho financeiro.

* **Loja 1**: R\$1.534.509,12
* **Loja 2**: R\$1.488.459,06
* **Loja 3**: R\$1.464.025,03
* **Loja 4**: R\$1.384.497,58 *(pior faturamento)*

#### 2. Média das Avaliações de Clientes por Loja

A satisfação do cliente é um fator importante para a fidelização. As médias foram:

* **Loja 1**: 3,98 *(pior avaliação)*
* **Loja 2**: 4,04
* **Loja 3**: 4,05
* **Loja 4**: 4,00 *(segunda pior avaliação)*

A **Loja 1 e Loja 2** receberam as menores médias de avaliação, indicando possíveis problemas de atendimento, entrega ou qualidade dos produtos.

#### 3. Frete Médio por Loja

O frete médio também foi avaliado, pois afeta tanto os custos quanto a demanda. Os valores foram:

* **Loja 1**: R\$34,69 *(frete mais caro)*
* **Loja 2**: R\$33,62
* **Loja 3**: R\$33,07
* **Loja 4**: R\$31,28 *(frete mais barato)*

O baixo custo de frete da **Loja 4** é outro ponto de atenção, podendo diminuir a margem de lucro, a média bem discrepante das demais alerta sobre uma baixa procura.

#### 4. Faturamento Mensal por Loja

Analisando um tópico extra, dos últimos meses que possuímos dados, a **Loja 4 e Loja 2** foram as que tiveram menor faturamento, destacando sua queda de faturamento nos meses de Janeiro à Maio de 2023.

---

### Conclusão

Após a análise de todos os indicadores, conclui-se que **a Loja 4 é a que apresenta o pior desempenho geral e, portanto, deve ser vendida**. Os principais motivos são:

* **Menor faturamento entre todas as lojas**
* **Segunda pior avaliação média dos clientes**,
* **Menor custo médio de frete**, indicando baixa procura que as demais
* **Baixo faturamento mensal no ano de 2023**, indicando sua queda em faturamento

Assim, a venda dessa loja se mostra como a decisão mais racional para otimizar os investimentos do Senhor João.
